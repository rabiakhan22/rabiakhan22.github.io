```python
import pandas as pd
import altair as alt

url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/ufo-scrubbed-geocoded-time-standardized-00.csv"

df = pd.read_csv(url, header=None, names=[
    'datetime', 'city', 'state', 'country', 'shape', 
    'duration_seconds', 'duration_text', 'comments', 
    'date_posted', 'latitude', 'longitude'
])

print(df.shape)
print(df.head())
print(df['shape'].value_counts().head(10))
print(df['state'].value_counts().head(10))
```

```python
df_us = df[df['country'] == 'us'].copy()
df_us = df_us.dropna(subset=['shape', 'state'])
df_us = df_us[df_us['shape'] != 'unknown']

print(df_us.shape)
print(df_us['shape'].value_counts().head(10))
```

```python
alt.data_transformers.disable_max_rows()

shape_counts = df_us['shape'].value_counts().reset_index()
shape_counts.columns = ['shape', 'count']

selection = alt.selection_point(fields=['shape'])

chart1 = alt.Chart(shape_counts).mark_bar().encode(
    x=alt.X('count:Q', title='Number of Sightings'),
    y=alt.Y('shape:N', sort='-x', title='UFO Shape'),
    color=alt.condition(selection, 
                        alt.Color('count:Q', scale=alt.Scale(scheme='teals')), 
                        alt.value('lightgray')),
    tooltip=['shape:N', 'count:Q']
).add_params(
    selection
).properties(
    title='UFO Sightings by Shape (US)',
    width=500,
    height=400
)

chart1
```

```python
state_counts = df_us['state'].value_counts().reset_index()
state_counts.columns = ['state', 'count']
top20 = state_counts.head(20)

chart2 = alt.Chart(top20).mark_bar().encode(
    x=alt.X('count:Q', title='Number of Sightings'),
    y=alt.Y('state:N', sort='-x', title='State'),
    color=alt.Color('count:Q', scale=alt.Scale(scheme='oranges'), title='Count'),
    tooltip=['state:N', 'count:Q']
).properties(
    title='Top 20 US States by UFO Sightings',
    width=500,
    height=400
)

chart2
```
